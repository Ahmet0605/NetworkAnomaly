```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
import time

# Zaman ölçüm başlat
seconds = time.time()

# Klasör oluşturma fonksiyonu
def folder(f_name):
    try:
        if not os.path.exists(f_name):
            os.makedirs(f_name)
    except OSError:
        print("The folder could not be created!")

# Klasör oluştur
folder("./feature_pics/")

# 'attacks' klasöründeki dosyaları al
csv_files = os.listdir("attacks")

# Sütun başlıkları
main_labels = [
    "Flow Duration", "Total Fwd Packets", "Total Backward Packets", "Total Length of Fwd Packets",
    "Total Length of Bwd Packets", "Fwd Packet Length Max", "Fwd Packet Length Min",
    "Fwd Packet Length Mean", "Fwd Packet Length Std", "Bwd Packet Length Max",
    "Bwd Packet Length Min", "Bwd Packet Length Mean", "Bwd Packet Length Std",
    "Flow Bytes/s", "Flow Packets/s", "Flow IAT Mean", "Flow IAT Std", "Flow IAT Max",
    "Flow IAT Min", "Fwd IAT Total", "Fwd IAT Mean", "Fwd IAT Std", "Fwd IAT Max",
    "Fwd IAT Min", "Bwd IAT Total", "Bwd IAT Mean", "Bwd IAT Std", "Bwd IAT Max",
    "Bwd IAT Min", "Fwd PSH Flags", "Bwd PSH Flags", "Fwd URG Flags", "Bwd URG Flags",
    "Fwd Header Length", "Bwd Header Length", "Fwd Packets/s", "Bwd Packets/s",
    "Min Packet Length", "Max Packet Length", "Packet Length Mean", "Packet Length Std",
    "Packet Length Variance", "FIN Flag Count", "SYN Flag Count", "RST Flag Count",
    "PSH Flag Count", "ACK Flag Count", "URG Flag Count", "CWE Flag Count",
    "ECE Flag Count", "Down/Up Ratio", "Average Packet Size", "Avg Fwd Segment Size",
    "Avg Bwd Segment Size", "Fwd Avg Bytes/Bulk", "Fwd Avg Packets/Bulk", "Fwd Avg Bulk Rate",
    "Bwd Avg Bytes/Bulk", "Bwd Avg Packets/Bulk", "Bwd Avg Bulk Rate", "Subflow Fwd Packets",
    "Subflow Fwd Bytes", "Subflow Bwd Packets", "Subflow Bwd Bytes", "Init_Win_bytes_forward",
    "Init_Win_bytes_backward", "act_data_pkt_fwd", "min_seg_size_forward", "Active Mean",
    "Active Std", "Active Max", "Active Min", "Idle Mean", "Idle Std", "Idle Max", "Idle Min", "Label"
]

# Sonuçları kaydetmek için bir dosya
ths = open("importance_list_for_attack_files.csv", "w")

# Her dosya için işlem
for j in csv_files:
    df = pd.read_csv(f"./attacks/{j}", usecols=main_labels)
    df = df.fillna(0)
    df["Label"] = [1 if label == "BENIGN" else 0 for label in df["Label"]]

    y = df["Label"].values
    del df["Label"]
    X = df.values

    X = np.float32(X)
    X[np.isnan(X)] = 0
    X[np.isinf(X)] = 0

    # Özellik önemlerini hesaplama
    forest = RandomForestRegressor(n_estimators=250, random_state=0)
    forest.fit(X, y)
    importances = forest.feature_importances_
    refclasscol = list(df.columns.values)

    # İlk 20 özellik için dataframe
    impor_bars = pd.DataFrame({'Features': refclasscol, 'Importance': importances})
    impor_bars = impor_bars.sort_values('Importance', ascending=False).head(20).set_index('Features')

    # İlk 5 özelliği al
    fea_ture = f"{j[:-4]}=[" + ",".join(f'"{feature}"' for feature in impor_bars.index[:5]) + "]"
    ths.write(fea_ture + "\n")

    # Ekrana yazdır
    print(f"{j[:-4]} importance list:")
    print(impor_bars.head(20))
    print(fea_ture)

    # Grafik oluştur ve ekranda göster
    plt.figure(figsize=(10, 5))
    impor_bars.plot.bar(color='skyblue')
    plt.title(f"{j[:-4]} Attack - Feature Importance")
    plt.ylabel('Importance')
    plt.tight_layout()
    plt.show()

    # Grafik kaydetme
    plt.savefig(f"./feature_pics/{j[:-4]}.pdf", bbox_inches='tight', format='pdf')
    plt.close()

    print("-----------------------------------------------------------------------------------------------")

# İşlemleri tamamla
print("\nAll attacks completed.")
print(f"Total operation time: {time.time() - seconds:.2f} seconds.")
ths.close()
```
